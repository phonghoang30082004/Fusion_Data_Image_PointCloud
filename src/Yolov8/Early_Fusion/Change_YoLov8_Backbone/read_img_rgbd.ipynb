```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

path = "/content/drive/MyDrive/Khóa Luận/Data_Object/Code/Early_Fusion_RGBD_Yolov8/4chan/rgbd_images_inpaint_training/000002_rgbd.png"

# 1) Đọc đủ 4 kênh (B,G,R,alpha)
rgbd = cv2.imread(path, cv2.IMREAD_UNCHANGED)
print("Shape:", rgbd.shape)  # (H, W, 4)

# 2) Tách riêng
bgr   = rgbd[..., :3]          # kênh màu gốc
alpha = rgbd[...,  3].astype(np.float32)  # depth (0–255)

# 3) Chuyển sang RGB để hiển thị với matplotlib
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

# Hiển thị với kích thước lớn hơn
plt.figure(figsize=(50, 15))
plt.subplot(1, 2, 1)
plt.title("RGB gốc (BGR→RGB)")
plt.axis("off")
plt.imshow(rgb)

plt.subplot(1, 2, 2)
plt.title("Depth (alpha channel)")
plt.axis("off")
plt.imshow(alpha, cmap="gray")
plt.show()
```
